```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# load in student_data
student_data = pd.read_csv("C:\\Users\\Ravi\\OneDrive\\Desktop\\DataCamp\\_data\\student-alcohol-consumption.csv")
```

```python
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", 
            row="study_time");
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",row = "famsup",
            row_order=["yes", "no"], col_order = ["yes","no"]);
```

```python
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]
```

```python
# Create a box plot and set the order of the categories
sns.catplot(y="G3",x="study_time", data = student_data, kind = "box", order = study_time_order);
```

```python
# Load in mpg data           
mpg = pd.read_csv("C:\\Users\\Ravi\\OneDrive\\Desktop\\DataCamp\\_data\\mpg.csv")
```

```python
# Make scatter plot
sns.relplot(x="acceleration", y="mpg", data = mpg, kind = "scatter", style='origin', hue = 'origin');
```

```python
# Make line plot
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line", ci = "sd");
```

```python
# Make a second line plot
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin",markers = True, dashes = False);
```
